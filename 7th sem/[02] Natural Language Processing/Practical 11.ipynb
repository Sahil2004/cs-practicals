```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline
```

```python
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment"
)
```

```python
texts = [
    "The movie was amazing and enjoyable",
    "The film was not good at all",
    "I expected more but it was okay",
    "Absolutely terrible experience",
    "Brilliant acting and great direction"
]
```

```python
results = sentiment_analyzer(texts)

for text, result in zip(texts, results):
    print(text)
    print("Sentiment:", result["label"], "| Confidence:", round(result["score"], 3))
    print("-" * 50)
```

```python
def map_sentiment(label):
    stars = int(label[0])
    return "Positive" if stars >= 4 else "Negative"

for text, result in zip(texts, results):
    print(text, "->", map_sentiment(result["label"]))
```
