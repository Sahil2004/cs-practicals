```python
from transformers import pipeline
from datasets import load_dataset
```

```python
dataset = load_dataset("amazon_polarity", split="test[:200]")
```

```python
dataset[0]
```

```python
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
for i in range(5):
    review = dataset[i]["content"]
    true_label = "Positive" if dataset[i]["label"] == 1 else "Negative"

    prediction = sentiment_analyzer(review[:512])[0]

    print("Review:", review[:200], "...")
    print("True Label:", true_label)
    print("Predicted:", prediction["label"], "| Confidence:", round(prediction["score"], 3))
    print("-" * 70)
```

```python
correct = 0
total = 50

for i in range(total):
    review = dataset[i]["content"]
    true = dataset[i]["label"]

    pred = sentiment_analyzer(review[:512])[0]["label"]
    pred_label = 1 if pred == "POSITIVE" else 0

    if pred_label == true:
        correct += 1

print("Accuracy on", total, "samples:", correct / total)
```

```python
user_review = input("Enter an Amazon-style product review: ")

output = sentiment_analyzer(user_review)[0]

print("\nPredicted Sentiment:", output["label"])
print("Confidence:", round(output["score"], 3))
```
