# Step 1: Data Pre-processing

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv("/content/CO22358_gearpartsales.csv")
X = dataset.iloc[:,[-3]].values
y = dataset.iloc[:,[-1]].values
```

## Handling the missing values

```python
from sklearn.impute import SimpleImputer

imputer=SimpleImputer(strategy = 'mean', missing_values = np.nan)

imputerX=imputer.fit(X[:,:])
X[:,:]=imputerX.transform(X[:,:])

imputerY=imputer.fit(y[:,:])
y[:,:]=imputerY.transform(y[:,:])
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
y_train = y_train.ravel()
y_test = y_test.ravel()
```

## Feature Scaling (Normalization) of data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler(with_mean=False)
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

# Step 2. Training of the SVR Model

## Fitting SVR to the dataset

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, y_train)
```

## Predicting the result

```python
y_predict = regressor.predict(X_test)
```

## MSE

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_predict)
print(f"Mean Squared Error (MSE): {mse}")
```

## R Squared Error

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_predict)
print(f"R-squared (R²): {r2}")
```

# Step 2: Training of the Decision Tree model

## Fitting the Decision Tree Regression to the dataset

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X_train, y_train)
```

## Predicting the result

```python
y_predict = regressor.predict(X_test)
```

## MSE

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_predict)
print(f"Mean Squared Error (MSE): {mse}")
```

## R Squared Error

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_predict)
print(f"R-squared (R²): {r2}")
```

# Step 2: Training of the Random Forest model

## Fitting the Random Forest Regression to the dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X_train, y_train)
```

## Predicting the result

```python
y_predict = regressor.predict(X_test)
```

## MSE

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_predict)
print(f"Mean Squared Error (MSE): {mse}")
```

## R Squared Error

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_predict)
print(f"R-squared (R²): {r2}")
```
